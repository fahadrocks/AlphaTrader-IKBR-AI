```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Download data
data = yf.download("TSLA", period="1y", interval="1d")

# Step 2: Creating a Function (Very Important for your AI)
def calculate_sma(data, w):
    """Calculate Simple Moving Average (SMA)
    SMA = Average of last 'window' days closing price
    This is one of the most basic but powerful technical indicators."""
    return data['Close'].rolling(window=w).mean()

# Step 3: Use the function
data['SMA_20'] = calculate_sma(data, 20)    # 20-day SMA
data['SMA_50'] = calculate_sma(data, 50)    # 50-day SMA

# Step 4: Plot everything
plt.figure(figsize=(14, 7))
plt.plot(data['Close'], label='TSLA Close Price', linewidth=2)
plt.plot(data['SMA_20'], label='20-day SMA', linewidth=2)
plt.plot(data['SMA_50'], label='50-day SMA', linewidth=2)

plt.title("TSLA Price with Simple Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()

# Step 5: Show recent data with indicators
print("Last 10 Days with SMAs:")
print(data[['Close', 'SMA_20', 'SMA_50']].tail(10))
```
