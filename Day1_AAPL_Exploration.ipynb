```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download real 6 months of Apple stock data (live market data!)
data = yf.download("AAPL", period="6mo", interval="1d")

# Plot a professional price chart
data[['Open', 'High', 'Low', 'Close']].plot(figsize=(12,6), title="AAPL Price - First Data Pull!")
plt.grid(True)
plt.show()

print(data.tail())   # Shows the most recent prices
```

### Options Theory Notes (Chapter 1 + first half of Chapter 2)

- **Call Option**: Gives the buyer the right (but not obligation) to BUY 100 shares at a fixed strike price before expiration.
- **Put Option**: Gives the buyer the right to SELL 100 shares at a fixed strike price.
- **Covered Call**: Own 100 shares + sell a call option against them. Generates steady premium income with limited risk — perfect strategy for beginners and steady income in the AI trading bot.
