```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']
data = yf.download(tickers, period='1y', interval='1d')

data= data.dropna()
data= data.round(2)

def sma(series, window):
    return series.rolling(window=window).mean()

plt.figure(figsize=(10, 5))

for ticker in tickers:
    plt.plot(data['Close'][ticker], label= f'{ticker} Close Price')
    plt.plot(sma(data['Close'][ticker], window=20), label= f'{ticker} 20-Day SMA')
plt.title('Closing Price Over the Last Year')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid()
plt.show()
```

### Options Theory Notes (Chapter 5)

- **Protective Put Strategy**: Owning shares of a stock and buying a put option to protect against a significant drop in price. It acts as insurance for your long position.

- **Covered Call Strategy**: Owning shares and selling a call option against them to generate premium income. It provides limited downside cushion but caps upside potential.

- **Key Difference**: 
  - Protective Put costs money (debit) but keeps unlimited upside.
  - Covered Call generates money (credit) but limits upside if the stock rises sharply.
